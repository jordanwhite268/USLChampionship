```python
import requests
from bs4 import BeautifulSoup

url = "https://fbref.com/en/comps/73/schedule/USL-Championship-Scores-and-Fixtures"
response = requests.get(url)

if response.status_code == 200:
    print("Request successful!")
else:
    print(f"Failed to fetch page. Status code: {response.status_code}")




```

```python
soup = BeautifulSoup(response.content, "html.parser")

# Check the entire page HTML structure to understand the table's structure
print(soup.prettify())  # This prints the HTML, which can help identify the right table and rows.
```

```python
# Look for the table using a different approach
table = soup.find("table", {"id": "schedule"})
if table:
    rows = table.find_all("tr")
    print(f"Found {len(rows)} rows.")
else:
    print("Table not found.")
```
