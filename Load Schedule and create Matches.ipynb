```python
import pandas as pd
import numpy 
#Read in excel file 
scheduledf= pd.read_csv(r'Data\2024Schedule.csv', delimiter = ",")
scheduledf.head(10)
```

```python
#drop non-playoff
value_to_drop = 'Conference Quarterfinals'
scheduledf = scheduledf[scheduledf["Round"] != value_to_drop]
value_to_drop = 'Round'
scheduledf = scheduledf[scheduledf["Round"] != value_to_drop]
value_to_drop = 'Conference Qu'
scheduledf = scheduledf[scheduledf["Round"] != value_to_drop]
scheduledf = scheduledf.dropna(subset=["Round"])
scheduledf.tail()
```

```python
# drop round, match report, extra columns.
scheduledf = scheduledf.drop(columns=["Round"])
scheduledf = scheduledf.drop(columns=["Wk"])
scheduledf = scheduledf.drop(columns=["Day"])
scheduledf = scheduledf.drop(columns=["Match Report"])
scheduledf = scheduledf.drop(columns=["Unnamed: 13"])
scheduledf = scheduledf.drop(columns=["Unnamed: 14"])
scheduledf = scheduledf.drop(columns=["Unnamed: 15"])
# add column Is_Playoff, assign values FALSE
scheduledf["Is_Playoff"] = False
#add column Match ID
scheduledf["Match_ID"] = range(1, len(scheduledf) + 1)
#set matchID to the index 
scheduledf.set_index("Match_ID", inplace=True)
#add column for season
scheduledf["Season"]=2024
#rename columns 
scheduledf.rename(columns={"Date": "Match_Date", "Time": "Match_Time_Local",
                           "Home": "Home_Team_ID", "Away": "Away_Team_ID"}, inplace=True)


scheduledf.head()
```

```python
#cast date to date type
scheduledf["Match_Date"] = pd.to_datetime(scheduledf["Match_Date"]).dt.date
#split time into EST and Local
# Define a function to process each row
def process_match_time(row):
    match_time_local = row["Match_Time_Local"]
    # Check if parentheses exist
    if "(" in match_time_local and ")" in match_time_local:
        # Extract time inside parentheses
        match_time_est = match_time_local.split("(")[-1].split(")")[0].strip()
        match_time_local = match_time_local.split("(")[0].strip()
    else:
        # No parentheses, set EST to Local
        match_time_est = match_time_local.strip()
    return match_time_local, match_time_est

# Apply the function and create new columns
scheduledf[["Match_Time_Local", "Match_Time_EST"]] = scheduledf.apply(
    lambda row: process_match_time(row), axis=1, result_type="expand"
)

# Convert columns to time dtype in 24-hour format
scheduledf["Match_Time_Local"] = pd.to_datetime(scheduledf["Match_Time_Local"], format="%H:%M").dt.time
scheduledf["Match_Time_EST"] = pd.to_datetime(scheduledf["Match_Time_EST"], format="%H:%M").dt.time

scheduledf.head()


```
